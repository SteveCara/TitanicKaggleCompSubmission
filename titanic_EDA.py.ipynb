```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import the data

train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
# view characteristics of training data

train_data.describe()
```

```python
# view all columns in training data

train_data.columns
```

```python
# view correlation between variables training data

train_data.corr()
```

```python
# create lmplot

# create palette
pal = dict(male="#6495ED", female="#F08080")

# set input data
g = sns.lmplot(x="Age", y="Survived", col="Sex", hue="Sex", data=train_data,
               palette=pal, y_jitter=.02, logistic=True, truncate=False)

g.set(xlim=(0, 80), ylim=(-.05, 1.05))

# from this we can infer:
# survival rate decreases with age for males
# survival rate increases with age for females
# more females survived than males
```

```python
# generate a heatmap to visualise correlations
corr = train_data.corr()

# draw the heatmap
cmap = sns.cubehelix_palette(light=1, as_cmap=True)

sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# joint plot

sns.jointplot(x = 'Age', y = 'Survived', data = train_data, kind = 'hex', color = '#4CB391')

# most of the survivers were aged ~ 18-35
```
